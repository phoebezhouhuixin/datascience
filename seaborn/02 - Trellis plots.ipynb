# Trellis plots

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fb_data=pd.read_csv('data/dataset_Facebook.csv',
                     sep=r'\s*;\s*', engine='python')
fb_data.head()
```

```python
fb_data.describe()
```

```python
g=sns.FacetGrid(fb_data, col="Type")
```

```python
g=sns.FacetGrid(fb_data, col="Type")
g.map(plt.hist, "Total Interactions")
```

```python
g=sns.FacetGrid(fb_data,col="Type")
g.map(plt.hist,"Total Interactions")

plt.xlim(-20,3000)
plt.ylim(0,30)
```

```python
g=sns.FacetGrid(fb_data, col="Type")
g.map(plt.scatter,"Lifetime Post Total Reach", "Lifetime Engaged Users")
```

```python
g=sns.FacetGrid(fb_data,col="Type",hue="Paid")
g.map(plt.scatter,"Lifetime Post Total Reach","Lifetime Engaged Users", alpha=0.7)
g.add_legend()
```

```python
g=sns.FacetGrid(fb_data, row="Category", col="Type", hue="Paid", margin_titles=True)
g.map(sns.regplot,"Post Month", "Total Interactions", fit_reg=False)
g.add_legend()
plt.ylim(0,2500)
```

```python
g=sns.FacetGrid(fb_data,row="Category", col="Type",hue="Paid", margin_titles=True,
                gridspec_kws={"width_ratios":[5, 3, 2, 2]})

g.map(sns.regplot,"Post Month","Total Interactions",fit_reg=False)

g.add_legend()
plt.ylim(0,2500)
```

```python
h={0:"red",1:"green"}
g=sns.FacetGrid(fb_data,row="Category", col="Type",hue="Paid", margin_titles=True,
                palette=h)

g.map(sns.regplot,"Post Month","Total Interactions",fit_reg=False)

g.add_legend()
plt.ylim(0,2500)
```

```python
#Creating a dictionary for markers and passing it to hue_kws variable
h={0:"red",1:"green"}
i={"marker":["^","v"]}

g=sns.FacetGrid(fb_data,row="Category", 
                col="Type",
                hue="Paid", 
                margin_titles=True,
                palette=h,
                hue_kws=i)


g.map(sns.regplot,"Post Month","Total Interactions",fit_reg=False)

g.add_legend()
plt.ylim(0,2500)
```

### Customizing

```python
g=sns.FacetGrid(fb_data, col="Post Month", col_wrap=3)
g.map(sns.barplot,"Post Weekday", "Page total likes",color="#aaff33", edgecolor="white", lw=.5)

g.fig.subplots_adjust(wspace=1, hspace=0.5);
```

### PairGrid, for pairwise relationships

```python
g = sns.PairGrid(fb_data,
                 vars=["comment", "like", "share"])

g.map(plt.scatter);
```

```python
g = sns.PairGrid(fb_data,
                 vars=["comment", "like", "share"])
g.map_offdiag(plt.scatter) 
g.map_diag(sns.kdeplot) 
```

```python
g = sns.PairGrid(fb_data,
                 vars=["comment", "like", "share"],
                 hue="Type")
g.map_offdiag(plt.scatter) 
g.map_diag(sns.kdeplot) 
g.add_legend()
```

```python
g = sns.PairGrid(fb_data,
                 vars=["comment", "like", "share"])

g.map_diag(sns.kdeplot) 
g.map_upper(plt.scatter) 
g.map_lower(sns.regplot)
```

```python
g = sns.PairGrid(fb_data,
                 x_vars=["comment", "like", "share"],
                 y_vars=["Post Month", "Post Weekday", "Post Hour"],
                 hue="Type")
g.map(plt.scatter) 
g.add_legend()
```
