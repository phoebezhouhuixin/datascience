# Handle missing values in data

```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
df = pd.DataFrame(data={
    'feature_1' : [np.nan, 3, 6, 9, 12, 15, np.nan],
    'feature_2' : [100, np.nan, 200, 300, np.nan, np.nan, 600],
    'feature_3' : [1000, 500, 2000, 3000, 4000, 6000, 8000],
})

df
```

```python
df.isnull() 
```

```python
df.isnull().sum()
```

## fill nans with fillna

```python
df.fillna(method='pad', limit=1)
    # pad / ffill: propagate last valid observation forward to next valid
    # backfill / bfill: use next valid observation to fill gap.
    # if method is specified, limit is the maximum number of consecutive
    # NaN values to forward/backward fill. In other words, if there is
    # a gap with more than this number of consecutive NaNs, it will only
    # be partially filled. If method is not specified, this is the
    # maximum number of entries along the entire axis where NaNs will be
    # filled. Must be greater than 0 if not None.
```

```python
#help(df.fillna)
```

```python
df.fillna(method='pad', limit=2)
```

```python
df.fillna(method = 'bfill')
```

```python
df.dropna(axis=0) # drop rows with nan
```

```python
df.dropna(axis=1) # drop columns with nan
```

Only drop columns which have at least 90% NaNs

```python
df.dropna(thresh=int(df.shape[0] * .9), axis=1)
```

## fillna with mean of Series

```python
df['feature_1'].fillna(df['feature_1'].mean())
```

```python
df['feature_2'].interpolate() # linear interpolation (equally spaced values between the nearest valid value before and the nearest valid value after)
```

## can also fill nans with replace()

```python
df.replace(np.nan, 0)
```

```python
boxplot = mall_customers_info.boxplot(grid=False, rot=45, figsize=(10, 8));
```

```python
boxplot = mall_customers_info.boxplot(grid=False, fontsize=15, 
                                      column=['Annual_Income'], figsize=(10, 8));
```

```python
mall_customers_info.loc[mall_customers_info['Annual_Income'] > 125]             
```

```python
Q1 = mall_customers_info['Annual_Income'].quantile(0.25)
Q3 = mall_customers_info['Annual_Income'].quantile(0.75)

print(Q1, Q3)
```

```python
IQR = Q3 - Q1
mall_customers_info['Annual_Income_Outlier'] = False
Q1 + 1.5 * IQR
```

```python
for index, row in mall_customers_info.iterrows():
    
    if row['Annual_Income']  > (Q1 + 1.5 * IQR):
        mall_customers_info.at[index, 'Annual_Income_Outlier'] = True
```

```python
mall_customers_info['Annual_Income_Outlier'].sum()
```

```python
non_outliers = mall_customers_info.loc[mall_customers_info['Annual_Income_Outlier'] == False]
non_outliers.head()
```

```python
mean = non_outliers['Annual_Income'].mean()
mean
```

```python
for index, row in mall_customers_info.iterrows():
    
    if row['Annual_Income_Outlier']  == True:
        mall_customers_info.at[index, 'Annual_Income'] = mean
```

```python
mall_customers_info.sample(5)
```

```python
boxplot = mall_customers_info.boxplot(grid=False, fontsize=15, 
                                      column=['Annual_Income'], figsize=(10, 8));
```

# Performing feature selection: univariate, RFE and Lasso

```python
house_data = pd.read_csv('data/HousingData.csv')
house_data.shape
```

```python
house_data.sample(10)
```

```python
col_names = ['CrimeRate', 'ZonedRatio', 'IndusRatio', 
             'AlongRiver', 'NO2Level', 'RoomsPerHouse', 
             'OldHomeRatio', 'DisFromCenter', 'RoadAccessIndex', 
             'PropTaxRate', 'PupilTeacherRatio', 'MedianHomeValue']
house_data.columns = col_names
house_data.sample(5)
```

```python
house_data.info()
```

```python
house_data = house_data.replace('NA', np.nan)
house_data.isnull().sum().sort_values(ascending=False)
```

```python
data = house_data.fillna(house_data.mean())
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.shape
```

```python
target = data['MedianHomeValue']
features = data.drop('MedianHomeValue', axis=1)
```

```python
features.shape, target.shape
```

* Univariate feature selection works by selecting the best features based on univariate statistical tests. It can be seen as a preprocessing step to an estimator.
* The methods based on F-test estimate the degree of linear dependency between two random variables. 
* Mutual information methods can capture any kind of statistical dependency, but being nonparametric, they require more samples for accurate estimation

```python
from sklearn.feature_selection import SelectKBest # SelectKBest removes all but the  highest scoring features
# SelectPercentile removes all but a user-specified highest scoring percentage of features

from sklearn.feature_selection import f_regression # This is a scoring function to be used in a feature selection procedure, not a free standing feature selection procedure.
# For regression: f_regression, mutual_info_regression
# For classification: chi2, f_classif, mutual_info_classif

select_univariate = SelectKBest(f_regression, k=5).fit(features, target)
feature_mask = select_univariate.get_support()
feature_mask
```

```python
features.columns[feature_mask]
```

```python
pd.DataFrame({'FeatureName': features.columns, 
              'Score': select_univariate.scores_}).sort_values(by='Score', 
                                                               ascending=False)
```

```python
uni_df = pd.DataFrame({'Univariate Method': features.columns[feature_mask]})
uni_df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

linear_regression = LinearRegression()
rfe = RFE(estimator = linear_regression, 
          n_features_to_select = 5, 
          step = 1)
rfe.fit(features,target)
```

```python
rfe.get_support()
```

```python
rfe_features = features.columns[rfe.support_]
rfe_features
```

```python
pd.DataFrame({'FeatureName': features.columns, 
              'Rank': rfe.ranking_}).sort_values(by='Rank')
```

```python
rfe_df = pd.DataFrame({'RFE Method': rfe_features})
rfe_df
```

## Regularization technique for feature selection 
* One of the most common mechanisms for avoiding overfit is called regularization. Regularized machine learning model, is a model whose loss function contains another element that should be minimized as well. 
* The first one is the one youve seen beforethe sum of distances between each prediction and its ground truth. The second element though, a.k.a the regularization term, punishes the loss function for high values of the coefficients.
* Ridge regression (L2) enforces the coefficients to be lower, but it does not enforce them to be zero. That is, it will not get rid of irrelevant features but rather minimize their impact on the trained model.
* Lasso regression (L1) is able to shrink some of the coefficients to zero. Therefore, that feature can be removed from the model.

```python
from sklearn.linear_model import  Lasso
lasso = Lasso(alpha=1.0)
lasso.fit(features, target)
```

```python
lasso_coef = pd.DataFrame({'Feature': features.columns, 
                           'LassoCoef': lasso.coef_}).sort_values(by = 'LassoCoef',
                                                                  ascending =False)
lasso_coef
```

```python
lasso_coef = pd.DataFrame({'Feature': features.columns, 'LassoCoef': abs(lasso.coef_)}).sort_values(by = 'LassoCoef', ascending =False).head()
lasso_coef
```

```python
lasso_df = lasso_coef["Feature"].reset_index()
lasso_df.drop("index", inplace = True, axis = 1)
lasso_df
```

```python
comp_selected_col_df = [uni_df, 
                        rfe_df, 
                        #back_df, 
                        #forw_df, 
                        lasso_df]

final_df = pd.concat(comp_selected_col_df, axis=1, ignore_index = True)
final_df
```
